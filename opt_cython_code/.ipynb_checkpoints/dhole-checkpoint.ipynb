```python
%%cython --annotate
import progressbar
from libc.math cimport sqrt, atan2, acos, abs
import numpy as np
from warnings import warn
from cython.parallel import prange
cimport cython

@cython.boundscheck(False) # turn off bounds-checking for entire function
@cython.wraparound(False)  # turn off negative index wrapping for entire function
@cython.cdivision(True)    # turn off check divission
def variogram(double[:,:] x, 
              double[:] v1, 
              double[:] v2, 
              int[:] constraint,
              int[:] restraint,
              float mindis=0, 
              float maxdis=1e10, 
              int maxnpairs =  90000000):
    
    # declare variables to make the code fast
    # ***************************************
    cdef int ndat, ndim, i, j, k, buff, ip
    cdef float xx[3]
    cdef float cdist, cazi, cpol
    
    ndat = x.shape[0]   
    
    assert x.shape[1]==3, '3D data required'
    assert x.shape[0]==v1.shape[0], 'Wrong array size'
    assert x.shape[0]==v2.shape[0], 'Wrong array size'
    assert x.shape[0]==constraint.shape[0], 'Wrong array size'
    assert x.shape[0]==restraint.shape[0], 'Wrong array size'
    assert ~np.isnan(x).any(), 'Nan values in coordinates'
    assert ~np.isnan(v1).any(), 'Nan values in variables'
    assert ~np.isnan(v2).any(), 'Nan values in variables'    
    
    # crate array as large as the number of pairs, same meaning gslib gamv
    if maxnpairs < ndat * (ndat-1)/2:
        warn ('Number of possible pairs may exeed maxnpairs')
    
    buff = min(maxnpairs, ndat * (ndat-1)/2)  #the numver of possible iterations is ndat * (ndat-1)/2
    
    #create output arrays
    dis_np,azi_np,pol_np, gam_np,cvar_np, pwgam_np, madgam_np, hm_np, tm_np, hv_np,tv_np = [np.zeros([buff], dtype=np.dtype(np.float32)) for i in range(11)]
    ii_np = np.zeros([buff], dtype=np.dtype(int))
    jj_np = np.zeros([buff], dtype=np.dtype(int))
    #create memory view
    cdef float [:] dis = dis_np
    cdef float [:] azi = azi_np
    cdef float [:] pol = pol_np
    cdef float [:] gam = gam_np
    cdef float [:] cvar = cvar_np
    cdef float [:] pwgam = pwgam_np
    cdef float [:] madgam = madgam_np
    cdef float [:] hm = hm_np
    cdef float [:] tm = tm_np
    cdef float [:] hv = hv_np
    cdef float [:] tv = tv_np
    cdef int [:] ii = ii_np
    cdef int [:] jj = jj_np
    
    # main loop to find pairs
    # ***************************************
    bar = progressbar.ProgressBar(ndat).start()
    ip = -1
    for i in range (ndat):
        bar.update(i+1)
        for j in range (i+1, ndat):
            # get spherical coordinates see https://mathworld.wolfram.com/SphericalCoordinates.html
            cdist = 0
            for k in range(3):
                xx[k] = x[i][k] - x[j][k]
                cdis+= xx[k]*xx[k]
                
            cdis = sqrt(cdis)
            cazi  = atan2(xx[1],xx[0])
            if cdis>0:
                cpol =  acos (xx[2]/cdis)
                
                
            # apply constraint (this is array of integers, pairs can be formed between points with the same constraint value)
            if constraint[i]!=constraint[j]: #prevent mixing data from different domain/drillhole
                continue
            
            if restraint[i]==restraint[j]: #prevent using data from same domain/drillhole
                continue
            
            #restrict distance
            if cdis< mindis:
                continue
            if cdis> maxdis:
                continue 
            
            
            ip+=1
            if ip>buff:
                continue
            
            ii[ip] = i                                  # pair source
            jj[ip] = j
            dis[ip] = cdis                              # dist
            azi[ip] = cazi                              # azimuth
            pol[ip] = cpol                              # dip angle
            gam[ip] = ((v1[i]-v1[j])*(v2[i]-v2[j]))/2   # variogram (note that works OK with two different variables)
            cvar[ip] = v1[i]*v2[j]                      # need to complete calculation outside 
            pwgam[ip] = ((v1[i]-v1[j])*(v2[i]-v2[j]))/((v1[i]+v1[j])/2*(v2[i]+v2[j])/2)
            madgam[ip] = abs((v1[i]-v2[j]))
            hm[ip] = v1[i]
            tm[ip] = v2[j]
            #hv[ip] 
            #tv[ip]
            
            
            
            # todo: calc  hm, tm, hv,tv, covariance, correlogram, general relative semivariogram, 
            #       pairwise relative semivar, semimadogram, see gslib book page 42
    ip+=1
    if ip>buff:
        warn ('Number of possible pairs exeeded maxnpairs and some pairs were ignored')
    
    return (ii_np[:ip], 
            jj_np[:ip], 
            dis_np[:ip],
            azi_np[:ip],
            pol_np[:ip], 
            gam_np[:ip],
            cvar_np[:ip],
            pwgam_np[:ip], 
            madgam_np[:ip], 
            hm_np[:ip], 
            tm_np[:ip], 
            hv_np[:ip],
            tv_np[:ip])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import pygslib
import numpy as np
```

```python
# read data
data = pd.read_csv('dataPT.csv')
# normal score 
data['dec_w'] = 1
ttable = pygslib.nonlinear.ttable(data['Au'],data['dec_w'])
data['Au_NS'] = pygslib.nonlinear.nscore(data['Au'], ttable[0], ttable[1], getrank=False)
```

```python
data
```

```python
x = data[['XPT', 'YPT', 'ZPT']].values
v1 = data['Au_NS'].values
constraint = data['Domain'].values.astype(int)
restraint = data['BHID'].values.astype(int)
n = constraint.shape[0]
constraint[np.random.randint(0,n,n//2)] = 0 # randonly set two region so we restrict number of pairs using all data points

# %timeit p_variogram(x, v1,v1, constraint)
ii, jj, dis,azi,pol, gam,cvar,pwgam, madgam, hm, tm, hv,tv = variogram(x, 
                                                                       v1,
                                                                       v1, 
                                                                       constraint, 
                                                                       restraint,
                                                                       maxdis = 50, 
                                                                       mindis = 40, 
                                                                       maxnpairs = 100)
```

```python
results = pd.DataFrame({'ii':ii, 
                        'jj':jj, 
                        'dis':dis,
                        'azi':azi,
                        'pol':pol, 
                        'gam':gam,
                        'cvar': cvar,
                        'pwgam': pwgam, 
                        'madgam':madgam, 
                        'hm':hm, 
                        'tm':tm, 
                        'hv':hv,
                        'tv':tv})

bsize = 1
results ['bin'] = (results ['dis']//bsize)*bsize # bin each 5m
```

```python
mask = ((results['dis']>0) &  (results['dis']<20)) &  ((results['azi']>-0.5) & (results['azi']<0.5)) 
covari=results.loc[mask, 'cvar'].sum()/mask.shape[0] - results.loc[mask, 'hm'].mean()*results.loc[mask, 'tm'].mean()
mdg=results.loc[mask, 'madgam'].mean()
fig, ax = plt.subplots(figsize=(8, 10))
results.loc[mask].plot.scatter(x='dis', y='gam', c='gam', cmap = 'viridis',  ax=ax)
results.loc[mask].groupby(by = 'bin')['gam'].mean().plot(marker = 'o')
plt.plot([0,20],[1,1], '--g', label = ' variance 1')
plt.legend()
plt.title('one lag')
plt.show()
```

```python
ax = results.loc[results['bin']==20, 'gam'].hist(density=True, bins = np.arange(0,3, 0.2))
results.loc[results['bin']==20, 'gam'].plot.kde(ax = ax, xlim=(0,3))
results.loc[results['bin']==20, 'gam'].mean()
```

```python
from sklearn.neighbors import KernelDensity
kde = KernelDensity(kernel='gaussian', bandwidth=0.75).fit(results.loc[results['bin']==10, ['gam']])
X_plot = np.linspace(0, 25, 1000)[:, np.newaxis]
log_dens = kde.score_samples(X_plot)
mean_exp = results.loc[results['bin']==10, 'gam'].mean()

# Mean and Variance - Monte Carlo see https://stackoverflow.com/questions/55788868/how-to-return-mean-value-or-expectation-value-of-a-distribution-estimated-via
n_samples = 10000000
samples = kde.sample(n_samples)

mean_kde = samples.mean()
variance_kde = samples.var()


fig, ax = plt.subplots()
ax.plot(X_plot[:, 0], np.exp(log_dens))
plt.title("Gaussian Kernel Density")

print (f' mean experimental {mean_exp}, from {mean_kde}, +/- {2*np.sqrt(variance_kde)}') ### need a more realistic interval
```

```python
import vtk
from vtk.util import numpy_support
import pygslib

cell_pro = {
    'ii':ii, 
    'jj':jj, 
    'dis':dis,
    'azi':azi,
    'pol':pol, 
    'gam':gam,
    'cvar':cvar,
    'pwgam':pwgam, 
    'madgam':madgam, 
    'hm':hm, 
    'tm':tm, 
    'hv':hv,
    'tv':tv}

vtkfields={}
for k in cell_pro.keys():
    vtkfields[k] = numpy_support.numpy_to_vtk(cell_pro[k])
    vtkfields[k].SetName(k)
    vtkfields[k].SetNumberOfComponents(1)

vtk_v1 = numpy_support.numpy_to_vtk(v1)
vtk_v1.SetName('v1')
vtk_v1.SetNumberOfComponents(1)
vtk_v2 = numpy_support.numpy_to_vtk(v1) #v1
vtk_v2.SetName('v2')
vtk_v2.SetNumberOfComponents(1)

# now we create a set of vtk points
points= vtk.vtkPoints()
npoints = x.shape[0]
points.SetNumberOfPoints(npoints)

bar = progressbar.ProgressBar(npoints).start()
for i in range(npoints):
    bar.update(i+1)
    points.SetPoint(i,x[i][0], x[i][1], x[i][2])

# now we create a set of lines representing the cores and
# a line container (a cell array)
line = vtk.vtkLine()
lines = vtk.vtkCellArray()
nlines = ii.shape[0]

# populate this data
n=-1
bar2 = progressbar.ProgressBar(nlines).start()
for l in range(nlines): 
    bar2.update(l+1)
    line.GetPointIds().SetId(0,ii[l])
    line.GetPointIds().SetId(1,jj[l])
    lines.InsertNextCell(line)

# Create a polydata to store everything in
linesPolyData = vtk.vtkPolyData()

# Add the points to the dataset
linesPolyData.SetPoints(points)

# Add the lines to the dataset
linesPolyData.SetLines(lines)

#add properties
for k in vtkfields:
    linesPolyData.GetCellData().AddArray(vtkfields[k])

linesPolyData.GetPointData().AddArray(vtk_v1)
linesPolyData.GetPointData().AddArray(vtk_v2)

# save data to VTK file
pygslib.vtktools.SavePolydata(linesPolyData, 'test.vtp')
```
